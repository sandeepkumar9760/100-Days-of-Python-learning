# Python supports 3 categories of data types:
- Basic types :- (Integer , float , complex , string and boolean)
- Container types :- (List , tuples, dictionary and sets)
- User-defined types :- Class

```python
#Integer

print(5)
print(50000000000000)
```

```python
#Float (decimal numbers)

print(5.60)
print(7.2)
print(0.10)
```

```python
# comples numbers (generally we dont use but useful when doing mathematical computations)

print(4+3j)
```

```python
#String (generally a line or sentance or word)

print("I am an example of string ")
print("Hello All the best")
```

```python
#boolean

print(True)
print(False)
```

```python
#List (its just an array according to other programming languages)
#we write list in square brackets []
print([1,2,3,4,5])
```

```python
#tuples (we write tuples in normal brackets)

print((1,2,3,4,5))
```

```python
#sets ({In curls brackets})

print({1,2,3,4,5})
```

```python
#dictionary (key value pairs in curls brackets {})

print({"Name":"Sandeep kumar","Age":22,"College":"Lovely professional university"})
```
