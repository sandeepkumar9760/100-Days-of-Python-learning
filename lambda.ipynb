

# Lambda Function in Python

## 1. What is a Lambda Function?

A **Lambda Function** is a small anonymous (nameless) function in Python.

* It can have **any number of arguments**.
* It can have **only one expression**.
* It automatically returns the result of that expression.
* It is mainly used for short and simple operations.

### Syntax

```python
lambda arguments: expression
```

### General Form

```python
lambda x: x + 10
```

Here,

* `lambda` → Keyword used to create an anonymous function.
* `x` → Input argument.
* `x + 10` → Expression that is evaluated and returned.

---

# 2. Simple Example

```python
square = lambda x: x ** 2

print(square(5))
```

### Output

```
25
```

### Explanation

* `x = 5`
* Expression becomes:

```python
5 ** 2
```

Result:

```
25
```

---

# 3. Lambda with Multiple Arguments

```python
add = lambda a, b: a + b

print(add(10, 20))
```

### Output

```
30
```

Explanation

```
a = 10
b = 20

10 + 20 = 30
```

---

# 4. Lambda for Multiplication

```python
multiply = lambda x, y: x * y

print(multiply(4, 6))
```

### Output

```
24
```

---

# 5. Lambda for Finding Maximum

```python
maximum = lambda a, b: a if a > b else b

print(maximum(12, 20))
```

### Output

```
20
```

### Explanation

This is similar to

```python
if a > b:
    return a
else:
    return b
```

---

# 6. Lambda vs Normal Function

## Normal Function

```python
def square(x):
    return x ** 2

print(square(5))
```

## Lambda Function

```python
square = lambda x: x ** 2

print(square(5))
```

Both produce

```
25
```

Difference:

* `def` creates a named function.
* `lambda` creates an anonymous function.

---

# 7. Lambda with map()

`map()` applies a function to every element in an iterable.

Example:

```python
numbers = [1, 2, 3, 4, 5]

result = list(map(lambda x: x * 2, numbers))

print(result)
```

### Output

```
[2, 4, 6, 8, 10]
```

Explanation

Each element is multiplied by 2.

```
1 → 2
2 → 4
3 → 6
4 → 8
5 → 10
```

---

# 8. Lambda with filter()

`filter()` selects elements that satisfy a condition.

Example

```python
numbers = [10, 15, 20, 25, 30]

even = list(filter(lambda x: x % 2 == 0, numbers))

print(even)
```

### Output

```
[10, 20, 30]
```

Explanation

Condition:

```python
x % 2 == 0
```

Only even numbers are returned.

---

# 9. Lambda with sorted()

Example

```python
students = [
    ("John", 85),
    ("Alice", 92),
    ("Bob", 78)
]

sorted_students = sorted(students, key=lambda x: x[1])

print(sorted_students)
```

### Output

```
[('Bob', 78), ('John', 85), ('Alice', 92)]
```

Explanation

```
x[1]
```

means sort using the second element (marks).

---

# 10. Lambda with max()

```python
students = [
    ("John", 85),
    ("Alice", 92),
    ("Bob", 78)
]

topper = max(students, key=lambda x: x[1])

print(topper)
```

### Output

```
('Alice', 92)
```

---

# 11. Lambda with min()

```python
students = [
    ("John", 85),
    ("Alice", 92),
    ("Bob", 78)
]

lowest = min(students, key=lambda x: x[1])

print(lowest)
```

### Output

```
('Bob', 78)
```

---

# 12. Lambda with reduce()

First import `reduce`.

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

total = reduce(lambda x, y: x + y, numbers)

print(total)
```

### Output

```
15
```

Explanation

```
1 + 2 = 3

3 + 3 = 6

6 + 4 = 10

10 + 5 = 15
```

---

# 13. Using Lambda for String Operations

```python
greet = lambda name: "Hello " + name

print(greet("Rahul"))
```

### Output

```
Hello Rahul
```

---

# 14. Lambda with Boolean Condition

```python
check = lambda age: "Adult" if age >= 18 else "Minor"

print(check(20))
print(check(15))
```

### Output

```
Adult
Minor
```

# 15. Real-Life Example

Suppose we have employee data.

```python
employees = [
    ("Amit", 50000),
    ("Neha", 70000),
    ("Riya", 60000)
]

employees.sort(key=lambda x: x[1])

print(employees)
```

### Output

```
[
('Amit', 50000),
('Riya', 60000),
('Neha', 70000)
]
```

Sorted according to salary.

# Advantages of Lambda Function

* Short and concise syntax.
* No need to define a separate function.
* Useful with `map()`, `filter()`, `reduce()`, and `sorted()`.
* Makes code cleaner for simple operations.
* Automatically returns the result.

---

# Limitations of Lambda Function

* Can contain only **one expression**.
* Cannot contain multiple statements.
* Not suitable for large or complex logic.
* Harder to debug compared to normal functions.

---

# When to Use Lambda?

Use lambda when:

* The function is small.
* It is used only once.
* You need a quick function inside `map()`, `filter()`, `sorted()`, or `reduce()`.

Use a normal `def` function when:

* The logic is complex.
* Multiple statements are required.
* The function will be reused many times.
