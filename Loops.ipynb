# Loops
In Python, loops are used to execute a block of code repeatedly.
- While loop
- For loop

# Syntax for while loop

- while condition:
- code

```python
#printing table of 5 with while loop
i = 1
while i<11:
    print(5*i)
    i+=1
```

# Syntax for - for loop

- for i in range ():
-  code

```python
for i in range(1,11):
    print(5*i)
    
```
