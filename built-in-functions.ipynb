```python
# ==========================================
# PYTHON BUILT-IN FUNCTIONS - REVISION SHEET
# ==========================================

# -----------------------------
# Type Conversion
# -----------------------------
a = int("10")
b = float("10.5")
c = str(100)
d = bool(1)
e = list("hello")
f = tuple([1, 2, 3])
g = set([1, 2, 2, 3])
h = dict(name="John", age=20)

print(a, b, c, d, e, f, g, h)

# -----------------------------
# Mathematical Functions
# -----------------------------
nums = [10, 20, 30, 40, 50]

print(sum(nums))
print(max(nums))
print(min(nums))
print(len(nums))
print(abs(-25))
print(round(3.14159, 2))
print(pow(2, 5))
print(divmod(17, 5))

# -----------------------------
# Sequence Functions
# -----------------------------
numbers = [5, 3, 8, 1]

print(sorted(numbers))
print(reversed(numbers))
print(enumerate(numbers))
print(list(enumerate(numbers)))

# -----------------------------
# Character Functions
# -----------------------------
print(chr(65))
print(ord('A'))

# -----------------------------
# Boolean Functions
# -----------------------------
print(all([True, True, False]))
print(any([False, False, True]))

# -----------------------------
# Zip Function
# -----------------------------
names = ["Alice", "Bob", "Charlie"]
marks = [90, 80, 85]

print(list(zip(names, marks)))

# -----------------------------
# Map Function
# -----------------------------
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x*x, nums))
print(squares)

# -----------------------------
# Filter Function
# -----------------------------
nums = [1, 2, 3, 4, 5, 6]
even = list(filter(lambda x: x % 2 == 0, nums))
print(even)

# -----------------------------
# Range
# -----------------------------
print(list(range(5)))
print(list(range(2, 10)))
print(list(range(2, 20, 3)))

# -----------------------------
# Input
# -----------------------------
# name = input("Enter your name: ")
# print(name)

# -----------------------------
# Type Checking
# -----------------------------
x = 10
print(type(x))
print(isinstance(x, int))

# -----------------------------
# ID Function
# -----------------------------
a = 100
print(id(a))

# -----------------------------
# Binary, Octal, Hexadecimal
# -----------------------------
print(bin(10))
print(oct(10))
print(hex(10))

# -----------------------------
# ASCII
# -----------------------------
print(ascii("Hello"))
print(ascii("₹"))

# -----------------------------
# String Formatting
# -----------------------------
name = "John"
age = 25

print(format(3.14159, ".2f"))
print(f"My name is {name} and I am {age} years old.")

# -----------------------------
# Eval (Be careful)
# -----------------------------
print(eval("5+10"))

# -----------------------------
# Slice
# -----------------------------
text = "Python Programming"
print(text[slice(0, 6)])

# -----------------------------
# Sorting with Key
# -----------------------------
words = ["apple", "kiwi", "banana", "pear"]
print(sorted(words, key=len))

# -----------------------------
# Help
# -----------------------------
# help(len)

# -----------------------------
# Object Functions
# -----------------------------
class Student:
    school = "ABC School"

    def __init__(self):
        self.name = "Alice"

s = Student()

print(hasattr(s, "name"))
print(getattr(s, "name"))
setattr(s, "name", "Bob")
print(s.name)
delattr(s, "name")
print(hasattr(s, "name"))

# -----------------------------
# Open File
# -----------------------------
# file = open("sample.txt", "w")
# file.write("Hello World")
# file.close()

# -----------------------------
# Iterators
# -----------------------------
nums = [10, 20, 30]
it = iter(nums)

print(next(it))
print(next(it))
print(next(it))

# -----------------------------
# Complex Numbers
# -----------------------------
c = complex(2, 3)
print(c)

# -----------------------------
# Frozenset
# -----------------------------
fs = frozenset([1, 2, 3, 4])
print(fs)

# -----------------------------
# Memory View
# -----------------------------
b = bytes([65, 66, 67])
mv = memoryview(b)
print(mv[0])

# -----------------------------
# Bytes
# -----------------------------
print(bytes([65, 66, 67]))

# -----------------------------
# Print
# -----------------------------
print("Hello", "World", sep="-", end="!\n")
```
