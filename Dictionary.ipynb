# **Dictionary**

*A dictionary in Python is a mutable, ordered collection of key–value pairs. Each key must be unique, and it is used to access its corresponding value.*

- Dictionary has no indexing.
- Dictionary is a mutable types.
- Keys -> immutable , values -> they can be mutable.
- Keys should be unique.


```python
mydict = {"Name":"Sandeep","Gender":"Male"}
print(type(mydict))
print(mydict)
```

```python
d2 = {"name":"rahul","name":"rohit"}
print(d2)
#it updates the value of the key as key is immutable.
```

```python
#2D dict

dict = {"Name":"Sandeep","Marks":{"S1":85,"S2":95}}
print(dict)

#like this you can deepdive into more dimension in dict.
```

```python
#Indexing
print(dict.keys())
print(dict.values())
print(dict["Name"])
```

```python
print(dict.items())
```

```python
#updating the dict 

dict.update({"School":"holy child school"})
print(dict)
```
