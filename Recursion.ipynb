# **Recursion**

*Recursion is a programming technique where a function calls itself to solve a smaller version of the same problem.*

- A recursive function has two important parts:

- Base case – Stops the recursion.
- Recursive case – Calls the function again with a smaller input.

```python
# Example 1: Factorial

# The factorial of a number is:

# 5! = 5 × 4 × 3 × 2 × 1 = 120
def factorial(n):
    if n == 0:          # Base case
        return 1
    return n * factorial(n - 1)   # Recursive case

print(factorial(5))
```
