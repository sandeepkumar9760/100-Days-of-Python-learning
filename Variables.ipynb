# VARIABLES IN PYTHON
- A variable in Python is a name used to store data in memory. It acts as a container that holds a value, which can be used and changed throughout a program.
- In python we dont have to tell the data type as it automatically gives the data type to the variable based on the input value.
- no variable declaration needed.

```python
name = "Sandeep kumar"
print(name)
```

```python
age = 21
print("age is:",age)
age = "sandeep"
print("age is:",age)

print("Dynamic binding - variables can store multiple data types in python")
```

```python
#variable declaration techniques

a=b=c=20
print(a,b,c)
a=6;b=7;c=8
print(a,b,c)
```
