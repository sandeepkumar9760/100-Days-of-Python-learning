| Function            | Purpose                        | Example                    | Output          |
| ------------------- | ------------------------------ | -------------------------- | --------------- |
| `len(s)`            | Length of string               | `len("hello")`             | `5`             |
| `lower()`           | Convert to lowercase           | `"HeLLo".lower()`          | `"hello"`       |
| `upper()`           | Convert to uppercase           | `"hello".upper()`          | `"HELLO"`       |
| `capitalize()`      | First letter capital           | `"python".capitalize()`    | `"Python"`      |
| `title()`           | Capitalize every word          | `"hello world".title()`    | `"Hello World"` |
| `strip()`           | Remove spaces both sides       | `" hello ".strip()`        | `"hello"`       |
| `lstrip()`          | Remove left spaces             | `" hello".lstrip()`        | `"hello"`       |
| `rstrip()`          | Remove right spaces            | `"hello ".rstrip()`        | `"hello"`       |
| `replace(old, new)` | Replace substring              | `"apple".replace("p","b")` | `"abble"`       |
| `find()`            | First occurrence index         | `"banana".find("na")`      | `2`             |
| `rfind()`           | Last occurrence index          | `"banana".rfind("na")`     | `4`             |
| `index()`           | Like find but errors if absent | `"abc".index("b")`         | `1`             |
| `count()`           | Count occurrences              | `"banana".count("a")`      | `3`             |
| `startswith()`      | Starts with?                   | `"hello".startswith("he")` | `True`          |
| `endswith()`        | Ends with?                     | `"hello".endswith("lo")`   | `True`          |
| `split()`           | Split into list                | `"a,b,c".split(",")`       | `['a','b','c']` |
| `'sep'.join(list)`  | Join list into string          | `"-".join(['a','b'])`      | `"a-b"`         |
| `splitlines()`      | Split by newlines              | `"a\nb".splitlines()`      | `['a','b']`     |
| `isalpha()`         | Only alphabets?                | `"Hello".isalpha()`        | `True`          |
| `isdigit()`         | Only digits?                   | `"123".isdigit()`          | `True`          |
| `isalnum()`         | Letters or digits?             | `"abc123".isalnum()`       | `True`          |
| `isspace()`         | Only spaces?                   | `"   ".isspace()`          | `True`          |
| `islower()`         | All lowercase?                 | `"hello".islower()`        | `True`          |
| `isupper()`         | All uppercase?                 | `"HELLO".isupper()`        | `True`          |
| `swapcase()`        | Toggle case                    | `"HeLLo".swapcase()`       | `"hEllO"`       |
| `center(width)`     | Center align                   | `"Hi".center(6,'-')`       | `"--Hi--"`      |
| `zfill(width)`      | Pad with zeros                 | `"42".zfill(5)`            | `"00042"`       |
