# Break 
- The break statement terminates the loop as soon as it is encountered.

Example:

```python
for i in range (1,11):
    if(i==6):
        break
    else:
        print(i)
```

# Continue 
- The continue statement skips the remaining code in the current iteration and moves to the next iteration.

Example:

```python
for i in range (1,11):
    if(i==6):
        continue
    else:
        print(i)
```

# Pass
- The pass statement does nothing. It is used as a placeholder when a statement is syntactically required but no action is needed.

Example:

```python
for i in range(1, 6):
    if i == 3:
        pass
    print(i)
```
