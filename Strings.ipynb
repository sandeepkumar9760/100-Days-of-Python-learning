

### **Strings are sequence of Characters**

In Python specifically, strings are a sequence of Unicode Characters.

* Creating Strings
* Accessing Strings
* Adding Chars to Strings
* Editing Strings
* Deleting Strings
* Operations on Strings
* String Functions


```python
#Creating 
c = 'Hello'
b = "hello"

print(c)
print(b)
```

```python
#multi line string 

c = '''hello my name is sandeep kumar.
I am from uttrakhand.
I studied in Lovely professional university.'''
print(c)
```

```python
#Assesing strings

#Concept of indexing
#Positive indexing
c='Hello'
print(c)
```

```python
print(c[0])
print(c[1])
print(c[2])
```

```python
print(c[7])
```

```python
#Negative indexing
print(c[-1])
print(c[-2])
print(c[-3])
print(c[-4])
print(c[-5])
```

```python
#Slicing

#print(str[start:end:steps])
#if you dont write steps so python takes step=1 as by default

print(c[0:4])
```

```python
print(c[:])
#this means all characters
```

```python
print(c[2:])
#this means reverse but till third last position
```

```python
print(c[::-1])
```

# **Editing and deleting strings in python**

```python
c[0]='s'
```

```python
#This means strings are immutable i.e you cant change strings in python
#you can completely change the string

c = "Sandeep"
print(c)
```

```python
#You can delete strings completely but not the specific alphabet from the string.
del c[0]
```

```python
#but you can delete whole string (like)

del c
print(c)
```
