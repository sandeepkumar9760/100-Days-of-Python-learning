# Operators
Operators are used to perform operations on variables and values. Python has the following operators.
- Arithmetic operators
- Comparison operators
- Logical operators
- Bitwise operators
- Assignment operators
- Identify operators
- Membership operators


```python
#Arithmetic Operators
x = 150
y = 10

print(x-y)
print(x+y)
print(x*y)
print(x/y)
print(x%y)
print(x**y)
print(x//y)
```

```python
#Comparison operators

print(x<y)
print(x<=y)
print(x>=y)
print(x>y)
```

```python
#Logical operators

x = True 
y = False

print(x and y)
print(x or y)
print(not x)
print(not y)
```

```python
#Bitwise operators (it works on binary values)

x = 2
y = 3

print("this is bitwiose and: ",x&y)
print("this is bitwise or: ",x|y)
print("this is left shift: ",x>>2)
print("this is right shift: ",y<<2)
```

```python
#Assignment operator

x = 250
print(x)

#=,+=,-= are the assignment operators
```

```python
#Identity operators

a=3
b=3
print(a is b)
```

```python
a = "hello"
b = 'hello'

print(a is b)
```

```python
a = [1,2,3]
b = [1,2,3]

print(a is b)
print("The output is not true because is operator sees the memory location if memory location is same then it gives True else false.")
```

```python
#Membership operators

a = [1,2,3,4,5]

print(4 in a)
print("Membership operators gives wheather there is occurence of the asked thing or not.")
print(10 not in a)
```
